```python
from pyaxidraw import axidraw
import vsketch
import random
import bezmerizing
```

```python
from numpy.random import uniform, normal, choice
import numpy as np
```

```python
from scipy.stats import truncnorm
def t_normal(a, b, mu, sigma):
    tn = truncnorm((a-mu)/sigma, (b-mu)/sigma, loc=mu, scale=sigma)
    return tn.rvs(1)[0]
```

```python
def uniform_glyph(n=10):
    pts = []
    for i in range(n):
        pts.append([uniform(-0.5, 0.5), uniform(-0.5, 0.5)])
    return pts
```

```python
def uniform_glyph_baseline(n=10):
    pts = [[-0.4, 0.5], [0.4, 0.5]]
    for i in range(n):
        pts.append([uniform(-0.5, 0.5), uniform(-0.5, 0.5)])
    return pts
```

```python
def grid_glyph():
    pts = []
    for i in range(int(t_normal(4, 12, 8, 2))):
        x = (choice([-0.5, 0, 0.5]))
        y = (choice([-0.5, 0, 0.5]))
        pts.append([x, y])
    return pts
```

```python
def make_scribble(width, height, steps, stddev=0):
    pts = []
    for i in range(steps):
        x = ((width / steps) * i) + normal(0, stddev)
        y = normal(0, height)
        pts.append([x, y])
    return pts
```

```python
vsk = vsketch.Vsketch()
vsk.size("300mm", "80mm")
vsk.scale("1mm")
pts = make_scribble(width=160, height=10, steps=60, stddev=5)
curve_pts = bezmerizing.Polyline(pts).catmull_spline(tightness=-0.5).to_polyline(10)
vsk.polygon(curve_pts)
vsk.display()
```

```python
vsk = vsketch.Vsketch()
vsk.size("200mm", "320mm")
vsk.scale("1mm")
for i in range(15):
    pts = make_scribble(width=20, height=5, steps=int(uniform(25, 100)), stddev=i * 0.5)
    curve_pts = bezmerizing.Polyline(pts).catmull_spline(tightness=uniform(-4, 1)).to_polyline(10)
    vsk.strokeWeight(choice([1, 2, 3]))
    with vsk.pushMatrix():
        vsk.translate(0, i * 20)
        vsk.polygon(curve_pts)
vsk.display()
```
